```python
# Cerinta 1

import pandas as pd

df = pd.read_csv('train.csv')

missing_values = df.isnull().sum()

duplicate_rows = df.duplicated().sum()

print(df.info())
print(f"Numarul de valori lipsa pentru fiecare coloana:\n{missing_values}")
print(f"Numarul de linii duplicate: {duplicate_rows}")
```

```python
# Cerinta 2

import matplotlib.pyplot as plt
import seaborn as sb
# Pentru cele 3 coloane:
# Ia numarul de valori pentru fiecare posibilitate,
# Imparte-le la cate sunt si inmulteste cu o suta

survived = df['Survived']
survival_rate = survived.value_counts(normalize=True) * 100

class_col = df['Pclass']
class_rate = class_col.value_counts(normalize=True) * 100

sex_col = df['Sex']
gender_rate = sex_col.value_counts(normalize=True) * 100

# 3 plot-uri pe un rand
main_plot, subplots = plt.subplots(1, 3, figsize=(12, 5))

# plot supravietuire
subplots[0].set_title('Procent de supraviețuire')
subplots[0].set_xticks([0, 1])
subplots[0].set_xticklabels(['N-au supravietuit', 'Au supraviețuit'])
sb.barplot(x=survival_rate.index, y=survival_rate.values, ax=subplots[0])

# plot pasageri pe clase
subplots[1].set_title('Procent pasageri pe clase')
subplots[1].set_xticks([0, 1, 2])
subplots[1].set_xticklabels(['Clasa 1', 'Clasa 2', 'Clasa 3'])
sb.barplot(x=class_rate.index, y=class_rate.values, ax=subplots[1])

# plot procente pe gen
subplots[2].set_title('Procentajul pe gen')
subplots[2].set_xticks([0, 1])
subplots[2].set_xticklabels(['Barbati', 'Femei'])
sb.barplot(x=gender_rate.index, y=gender_rate.values, ax=subplots[2])
plt.show()
```

```python
# Cerinta 3

# vector cu coloanele ce au ca tip valori numerice
only_numerical_cols = df.select_dtypes(include=['int64', 'float64'])
name_of_cols = only_numerical_cols.columns
df[name_of_cols].hist(bins=40, figsize=(20, 10))
plt.show()
```

```python
# Cerinta 4

cols_without_values = missing_values[missing_values > 0]
nr_linii = df.shape[0]
percentages = (cols_without_values / nr_linii) * 100

print(f'Coloanele ce au valori lipsa si cate din acestea sunt:\n{cols_without_values}')

print(f'Procentul acestora din numarul total:\n{percentages}')
```

```python
# Cerinta 5

import numpy as np
df = pd.read_csv('train.csv')
intervals = [(0, 20), (21, 40), (41, 60), (61,1000)]
l = []
age_col = df['Age']
for i in range(len(intervals)):
    interval_count = age_col[(age_col >= intervals[i][0]) & (age_col <= intervals[i][1])].count()
    l.append(interval_count)
Age_interval = []
age = df['Age']
for i in range(df.shape[0]):
    x = age[i]
    if pd.isna(x):
        # -1 daca nu exista varsta
        Age_interval.append(-1)
        continue
    for j in range(0, len(intervals)):
        if (intervals[j][0] <= x <= intervals[j][1]):
            idx_interval = int(j)
            break
    Age_interval.append(idx_interval)
df.insert(2, "Age_Interval", Age_interval, True)
x = np.array(['0 - 20', '21 - 40', '41 - 60', '   61 - max'])
y = np.array(l)
plt.bar(x, y)
plt.xlabel('Age intervals')
plt.ylabel('Number of people')
plt.title('Number of people in each age interval')
plt.show()
df.to_csv('train_with_added_col.csv', index=False)
```

```python
# Cerinta 6

# barbatii care au supravietuit
survived_males = df[(df['Sex'] == 'male') & (df['Survived'] == 1)]
nr_of_males = df[df['Sex'] == 'male'].shape[0]
intervals = [(0, 20), (21, 40), (41, 60), (61,1000)]
male_survivors_percentages= []
for i in range(len(intervals)):
    interval_count = survived_males[survived_males['Age_Interval'] == i].shape[0]
    male_survivors_percentages.append(interval_count / nr_of_males * 100)
x = np.array(['0 - 20', '21 - 40', '41 - 60', '   61 - max'])
y = np.array(male_survivors_percentages)
plt.bar(x, y)
plt.xlabel('Age intervals')
plt.ylabel('Percentage (%)')
plt.title('Male survival percentage based on age interval')
plt.show()
```

```python
# Cerinta 7

children = df[df['Age'] < 18]
adults = df[df['Age'] >= 18]

children_percentage = (children.shape[0] / df.shape[0]) * 100
print(f'Procentajul copiilor aflati la bord din totalul pasagerilor este: {round(children_percentage, 2)}%')

# coloana de supravietuire a copiilor
s_children = children['Survived']

# procent = nr copii supravietuitori / nr copii * 100
children_survival_rate = s_children[s_children == 1].count() / children.shape[0] * 100

# coloana de supravietuire a adultilor
s_adults = adults['Survived']

# procent = nr adulti supravietuitori / nr adulti * 100
adults_survival_rate = s_adults[s_adults == 1].count()/ adults.shape[0] * 100

x=['Children', 'Adults']
y=[children_survival_rate, adults_survival_rate]

plt.bar(x,y)
plt.title('Survival rate for children and adults')
plt.xlabel('Category')
plt.ylabel('Percent (%)')
plt.show()
```

```python
# Cerinta 8

# grupam in 2 categorii (0 sau 1 din coloana Survived) si
# si calculam pentru  fiecare media dupa age
medii_grupe = df.groupby('Survived')['Age'].transform('mean')
df['Age'] = df['Age'].fillna(medii_grupe)

# grupam in 2 categorii (0 sau 1 din coloana Survived) si
# fiecare grup primeste cea mai frecventa valoare din coloana cabin
cabina_frecventa = df.groupby('Survived')['Cabin'].transform(lambda x: x.mode()[0])
df['Cabin'] = df['Cabin'].fillna(cabina_frecventa)

# grupam in 2 categorii (0 sau 1 din coloana Survived) si
# fiecare grup primeste cea mai frecventa valoare din Embarked
port_frecventa = df.groupby('Survived')['Embarked'].transform(lambda x: x.mode()[0])
df['Embarked'] = df['Embarked'].fillna(port_frecventa)

df.to_csv('train_filled.csv', index=False)
```

```python
# Cerinta 9

sdf = df['Name'].str.extract('([A-Za-z]+\.)')

titluri_unice = sdf[0].unique().tolist()
# titlurile unice:
# ['Mr.', 'Mrs.', 'Miss.', 'Master.', 'Don.', 'Rev.', 'Dr.', 'Mme.', 'Ms.', 'Major.', 'Lady.', 'Sir.', 'Mlle.', 'Col.', 'Capt.', 'Countess.', 'Jonkheer.']

# genurile corespounzatoare:
genuri = ['male', 'female', 'female', 'male', 'male', 'male', 'male', 'female', 'female', 'male', 'female', 'male', 'female', 'male', 'male', 'female', 'male']

# lista initializata la 0 pentru fiecare titlu
titluri_contor = [0] * len(titluri_unice)

# itereaza prin toate liniile
for i in range(len(df)):
    # titlul si sexul persoanei curente
    titlu = sdf[0].iloc[i]
    sex = df['Sex'].iloc[i]
    # luam indexul din lista de titluri
    idx = titluri_unice.index(titlu)
    # daca sexul titlului corespunde cu cel al persoanei
    if genuri[idx] == sex:
        # creste contorul titlului respectiv
        titluri_contor[idx] += 1

midpoint = len(titluri_unice) // 2
parte_1_titluri = titluri_unice[:midpoint]
parte_2_titluri = titluri_unice[midpoint:]

parte_1_contor = titluri_contor[:midpoint]
parte_2_contor = titluri_contor[midpoint:]

plt.bar(parte_1_titluri, parte_1_contor)
plt.title('Cate persoane au un titlu anume')
plt.xlabel('Titluri')
plt.ylabel('Numar persoane')
plt.show()

print('\n')
plt.bar(parte_2_titluri, parte_2_contor)
plt.title('Cate persoane au un titlu anume')
plt.xlabel('Titluri')
plt.ylabel('Numar persoane')
plt.show()
```

```python
# Cerinta 10

# singur si a supravietuit
d1 = ((df['SibSp'] + df['Parch'] == 0) & (df['Survived'] == 1)).astype(int)
nr_singuri_viata = d1.sum()
rata_supravietuire_singuri = round(nr_singuri_viata / d1.shape[0] * 100, 2)

# acompaniat si a supravietuit
d2 = ((df['SibSp'] + df['Parch'] > 0) & (df['Survived'] == 1)).astype(int)
nr_acompaniati_viata = d2.sum()
rata_supravietuire_acompaniati = round(nr_acompaniati_viata / d2.shape[0] * 100, 2)

x = ['Singuri','Acompaniati']
y = [rata_supravietuire_singuri, rata_supravietuire_acompaniati]
plt.bar(x, y)
plt.title('Rata supravietuire in functie de stare se singuratate')
plt.ylabel('Procent (%)')

plt.show()

# primele 100 linii
first100 = df.head(100)
sb.catplot(data = first100, kind = 'swarm', x = 'Pclass', y = 'Fare', hue = 'Survived', height = 6, aspect = 2)
plt.title('Relatia dintre tarif, clasa si starea de supravietuire pentru primele 100 de inregistrari')
plt.show()
```
